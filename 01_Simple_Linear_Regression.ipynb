## Step 1: Import Required Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

## Step 2: Import the Dataset

```python
# Importing the dataset
dataset = pd.read_csv("https://raw.githubusercontent.com/aimlresearcher/Dataset-ML-Class/refs/heads/main/salary_data.csv")  # Use raw string to avoid escape character issues

# Separating Independent and Dependent Variables
X = dataset.iloc[:, :-1].values  # Independent variable (Years of Experience)
y = dataset.iloc[:, -1].values   # Dependent variable (Salary)
```

## Step 3: Splitting the Dataset

```python
# Splitting the dataset into Training and Test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=0)

# Printing the training and test sets (Optional)
print("X_train:\n", X_train)
print("y_train:\n", y_train)
print("X_test:\n", X_test)
print("y_test:\n", y_test)
```

## Step 4: Training the Simple Linear Regression Model

```python
# Training the Simple Linear Regression Model
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

## Step 5: Predicting the Test Set Results

```python
# Predicting the Test Set Results
y_pred = regressor.predict(X_test)

# Printing actual vs predicted values (Optional)
print("Actual y_test values:\n", y_test)
print("Predicted y_pred values:\n", y_pred)
```

## Step 6: Visualizing the Results

```python
# Visualizing the Training Set Results
plt.scatter(X_train, y_train, color='red', label="Actual Data")  # Scatter plot for actual data
plt.plot(X_train, regressor.predict(X_train), color='blue', label="Regression Line")  # Regression line
plt.title('Salary vs Experience (Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()

# Visualizing the Test Set Results
plt.scatter(X_test, y_test, color='red', label="Actual Data")  # Scatter plot for actual test data
plt.plot(X_train, regressor.predict(X_train), color='blue', label="Regression Line")  # Same regression line
plt.title('Salary vs Experience (Test Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()

# Visualizing Test Set Results using y_pred
plt.scatter(X_test, y_test, color='red', label="Actual Data")
plt.plot(X_test, y_pred, color='green', label="Predicted Salary")  # Line for predictions
plt.title('Salary vs Experience (Predictions)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

## Step 7: Making a Single Prediction

```python
# Making a single prediction (e.g., salary of an employee with 12 years of experience)
single_prediction = regressor.predict([[12]])
print("Predicted Salary for 12 years of experience:", single_prediction[0])
```

## Step 8: Getting Model Coefficients

```python
# Getting the Coefficient and Intercept
coefficient = regressor.coef_
intercept = regressor.intercept_

print("Coefficient (Slope):", coefficient[0])
print("Intercept:", intercept)
```

## Step 9: Manual Prediction (Extra Practice)

```python
# Manual prediction using the equation: y = mx + c
manual_prediction = intercept + coefficient[0] * 15
print("Manual Prediction for 15 years of experience:", manual_prediction)

# Auto prediction using the trained model
auto_prediction = regressor.predict([[15]])
print("Auto Prediction for 15 years of experience:", auto_prediction[0])
```

### Final Cleaned and Corrected Code

```python
# Importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Importing the dataset
dataset = pd.read_csv(r"https://raw.githubusercontent.com/aimlresearcher/Dataset-ML-Class/refs/heads/main/salary_data.csv")
X = dataset.iloc[:, :-1].values  # Independent variable
y = dataset.iloc[:, -1].values   # Dependent variable

# Splitting the dataset into Training and Test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=0)

# Training the Simple Linear Regression model
regressor = LinearRegression()
regressor.fit(X_train, y_train)

# Predicting the Test set results
y_pred = regressor.predict(X_test)

# Visualizing the Training Set Results
plt.scatter(X_train, y_train, color='red', label="Actual Data")
plt.plot(X_train, regressor.predict(X_train), color='blue', label="Regression Line")
plt.title('Salary vs Experience (Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()

# Visualizing the Test Set Results
plt.scatter(X_test, y_test, color='red', label="Actual Data")
plt.plot(X_train, regressor.predict(X_train), color='blue', label="Regression Line")
plt.title('Salary vs Experience (Test Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()

# Visualizing Test Set Predictions
plt.scatter(X_test, y_test, color='red', label="Actual Data")
plt.plot(X_test, y_pred, color='green', label="Predicted Salary")
plt.title('Salary vs Experience (Predictions)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()

# Making a single prediction (e.g., salary for 12 years of experience)
single_prediction = regressor.predict([[12]])
print("Predicted Salary for 12 years of experience:", single_prediction[0])

# Getting the Coefficient and Intercept
coefficient = regressor.coef_
intercept = regressor.intercept_

print("Coefficient (Slope):", coefficient[0])
print("Intercept:", intercept)

# Manual prediction using the equation: y = mx + c
manual_prediction = intercept + coefficient[0] * 15
print("Manual Prediction for 15 years of experience:", manual_prediction)

# Auto prediction using the trained model
auto_prediction = regressor.predict([[15]])
print("Auto Prediction for 15 years of experience:", auto_prediction[0])
```
